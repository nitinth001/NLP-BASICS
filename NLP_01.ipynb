Stemming (NLP) — Important Points (Short)

Definition:
Stemming is a process in NLP that reduces words to their root/base form by removing suffixes.

```python
words = ["writing","writes","programming","programs","history",
         "finally","finalized","eating","eats","eaten"]
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
for word in words:
  print(word + "->" + ps.stem(word))
```

```python
ps.stem("congratulations")
```

```python
ps.stem("sitting")
```

```python
#ways to improve porter stemming:
# REGEXP STEMMER:
```

RegexpStemmer (Regular Expression Stemmer) – NLTK

RegexpStemmer is a stemmer in NLTK that removes suffixes using regular expressions (regex patterns).

Unlike Porter or Lancaster stemmers, it does not use linguistic rules.
Instead, it removes parts of words based on a pattern you define.

```python
from nltk.stem import RegexpStemmer
reg_stemmer  = RegexpStemmer('ing$|s$|e$|able$', min=4)
reg_stemmer.stem('eating')
```

```python
from nltk.stem import RegexpStemmer
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
reg_stemmer.stem('ingeating') #ing$ means remove 'ing' only from last..
```

Snowball Stemmer (NLP)

Snowball Stemmer is an improved version of the Porter Stemmer used for reducing words to their base/stem form.

Snowball Stemmer removes suffixes from words using improved linguistic rules.

Important Features:

Improved Porter algorithm

Supports many languages

More consistent stemming

Faster and cleaner implementation

```python
from nltk.stem import SnowballStemmer
print(SnowballStemmer.languages)
```

```python
ss = SnowballStemmer('english')
ss.stem('eating')
```

Lemmatization (NLP)

Lemmatization is the process of converting a word to its base or dictionary form (called lemma) using vocabulary and grammar rules.

**Advantages:**

Produces real dictionary words

More accurate meaning

Useful in chatbots and language understanding

**Disadvantages:**

Slower than stemming

Requires dictionary (WordNet)

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
```

```python
lemmatizer.lemmatize('going',pos = 'n') # pos represents noun(n), verb(v), adverb(r), adjective(a) so based on word we have we select it
```

```python
lemmatizer.lemmatize('going',pos = 'v')
```

```python
words = ["writing","writes","programming","programs","history",
         "finally","finalized","eating","eats","eaten"]
for word in words:
  print(word + "->" + lemmatizer.lemmatize(word,pos='n'))
```

```python
words = ["writing","writes","programming","programs","history",
         "finally","finalized","eating","eats","eaten"]
for word in words:
  print(word + "->" + lemmatizer.lemmatize(word,pos='v'))
```

```python
lemmatizer.lemmatize('fairly',pos='v'), lemmatizer.lemmatize('sportingly',pos='v')
```

**STOPWORDS:**

Stopwords are very common words in a language that usually do not carry significant meaning in text analysis.
They are often removed during text preprocessing to focus on important words.

Important Points

Stopwords are language dependent

They do not add much meaning to text analysis

Removing them is called stopword removal

Used in NLP preprocessing

```python
paragraph = """
 Natural Language Processing is an important field of artificial intelligence that focuses on the interaction
 between computers and human language. In many real world applications, machines need to understand
 large amounts of text data in order to extract useful information from it. For example, companies analyze
 customer reviews to determine whether their products are liked or disliked by users. Similarly, search engines
 process millions of documents to return the most relevant results to a user's query. However, raw text data
 is often not in a format that is easy to analyze.
 These words are known as stopwords and they are usually removed during the preprocessing stage of natural language processing.
 By eliminating these unnecessary words, the size of the text data becomes smaller and the important keywords become easier to identify.
 This process helps machine learning models focus on the most meaningful parts of the text, thereby improving the
 efficiency and accuracy of tasks such as text classification, sentiment analysis, and information retrieval.
"""
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
stopwords.words('french')
```

```python
stopwords.words('spanish')
```

```python
stemmer = PorterStemmer()
sentences = nltk.sent_tokenize(paragraph)
```

```python
type(sentences)
```

```python
#apply stopwords and filter and then apply porter stemming:
for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
  sentences[i] = ' '.join(words) # converting all list fo wrods into sentences..
```

```python
sentences
```

```python
#uing snowball stemming now:
from nltk.stem import SnowballStemmer
ss= SnowballStemmer('english')
```

```python
for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  words = [ss.stem(word) for word in words if word not in set(stopwords.words('english'))]
  sentences[i] = ' '.join(words)
```

```python
sentences
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  words = [lemmatizer.lemmatize(word.lower(),pos ='v') for word in words if word not in set(stopwords.words('english'))]
  sentences[i] = ' '.join(words)
```

```python
sentences
```

**POS(PARTS OF SPEECH) TAGGING :**

Part of Speech (POS) refers to the grammatical category of a word in a sentence based on its function.
In NLP, POS tagging means assigning each word a label such as noun, verb, adjective, etc.

Why POS Tagging is Important

Helps understand sentence structure

Used in machine translation

Helps in lemmatization

Used in information extraction

Important for chatbots and NLP models

```python
paragraph= """
The quick brown fox jumps over the lazy dog
Artificial intelligence is transforming modern technology
She completed the project very quickly
Students are learning natural language processing today
The scientist discovered a new method for data analysis
Machine learning models require large amounts of data
The boy is playing football in the garden
"""
```

```python
sentences = nltk.sent_tokenize(paragraph)
sentences
```

```python
import nltk
nltk.download('averaged_perceptron_tagger_eng')
```

```python
#finding pos_tags:
for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  words = [word for word in words if word not in set(stopwords.words('english'))]
  pos_tag = nltk.pos_tag(words)
  print(pos_tag)
```
